#Começando o desafio com analise e tratamento de dados

```python
#Importando as bibliotecas importantes para o momento.

import pandas as pd
import numpy as np
```

```python
#Leitura dos nosso dados.

dados = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json')
```

```python
dados.head(7)
```

```python
#Substituir os nomes das colunas.

dados.rename(columns={'customerID':'ID',
                      'Churn':'Evasão',
                      'customer':'Cadastrais',
                      'phone':'Telefone',
                      'internet':'serviço_net',
                      'account':'Conta'}, inplace=True)
```

```python
dados
```

```python
#Vamos analisar todos os dados para que possamos fazer uma limpeza de dados nos campos.
dados.info()
```

```python
dados.isnull().sum()
```

```python
dados.shape
```

#Limpeza e tratamento do dataframe

```python
dados.head(2)
```

```python
dados_util = dados.drop(columns=['Cadastrais', 'Telefone', 'serviço_net', 'Conta'])
dados_util
```

```python
#Vamos agrupar as colunas para que tenha uma melhor visualização dos dados.

dados_cadastro = pd.json_normalize(data=dados['Cadastrais'], sep='_')
dados_telefone = pd.json_normalize(data=dados['Telefone'], sep='_')
dados_net = pd.json_normalize(data=dados['serviço_net'], sep='_')
dados_conta = pd.json_normalize(data=dados['Conta'], sep='_')
```

```python
dados_conta
```

```python
dados_att = pd.concat([dados_util, dados_cadastro, dados_telefone,dados_net, dados_conta], axis=1)
dados_att
```

```python
#Nos precisaremos de novo traduzir, já que adaptamos nossa coluna ao nossos pontos.

dados_att.rename(columns={'gender':'Genero',	"SeniorCitizen":'Idoso',
                          'Partner':'Parceiro','Dependents':'Dependentes',
                          'tenure':'dominios',	'PhoneService':'ServiçosCel',	
                          'MultipleLines':'MultiplasLinhas',
                          'InternetService':'ServiçoNet',
                          'OnlineSecurity':'SegurançaNet',	
                          'OnlineBackup	':'Backup',
                          'DeviceProtection':'ProteçãoDisp',
                          'TechSupport':'SuporteTec','StreamingTV':'TransmissaoTv',	
                          'StreamingMovies':'TransmissaoFilmes',
                          'Contract':'Contrato','PaperlessBilling':'FaturaFisica',	
                          'PaymentMethod':'MetodoPagamento',	'Charges_Monthly':'RecargaMês',
                          'Charges_Total':'RecargaTotal'}, inplace=True)
```

```python
dados_att
```

```python
#Vamos traduzir nossas linhas para que todos que não entendam de inglês possam analisar também.

#M = PAGAMENTO MANUAL / A = PAGAMENTO AUTOMATICO 

tradução = {'Yes':'Sim',
            'No':'Nao',
            'Male':'Homem',
            'Female':'Mulher',
            'Fiber optic':'FibraOtica',
            'No phone service':'Sem contrato',
            'One year':'Anual',
            'month-to-month':'Mensal',
            'Two year':'DoisAnos',
            'Mailed check':'PagamentoFisico(M)',
            'Electronic check':'PagamentoEletronico(M)',
            'Bank transfer (automatic)':'TransfBancaria(A)',
            'Credit card (automatic)':'CartãoCredito(A)'
            }

dados_att= dados_att.replace(tradução)

print(tradução)
```

```python
dados_att
```

```python
#Vamos considerar que as pessoas com número 0 não sejam idosos e 1 sejam idosos.
dados_att['Idoso'].value_counts()
```

```python
dados_idoso = {0:'não',
               1:'sim'}

dados_att['Idoso']= dados_att['Idoso'].replace(dados_idoso)

print(dados_idoso)
```

#Agora que acabamos nossa limpeza e transformação dos dados, vamos procurar e tratar dados ineficazes de nos ajudar.

```python
dados_att.info()
```

```python
#Nossa coluna RecargaTotal tem que ser númerica, então.

dados_att['RecargaTotal'] = dados_att['RecargaTotal'].str.replace(',', '.')
dados_att['RecargaTotal']
```

```python
dados_att['RecargaTotal'].isnull().sum()
```

```python
dados_att['RecargaTotal'] = pd.to_numeric(dados_att['RecargaTotal'],errors = 'coerce')
```
